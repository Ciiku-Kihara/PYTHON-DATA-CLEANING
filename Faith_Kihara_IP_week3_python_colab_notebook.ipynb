**Importing libraries**

```
# Importing the pandas and numpy libraries
#
import pandas as pd
import numpy as np
```

**Loading all our datasets**

```
# Loading the cells_geo dataset
#
cells_geo = pd.read_csv('cells_geo.csv')
cells_geo.head()

```

```
# Loading the first CDR dataset and getting a 
# a preview of its layout
#
url1 = "http://bit.ly/Telcom_dataset1"

cdr507 = pd.read_csv(url1)
cdr507.head()
```

```
# Loading the second CDR dataset and getting a 
# a preview of its layout
#
url2 = "http://bit.ly/Telcom_dataset2"

cdr508 = pd.read_csv(url2)
cdr508.head()
```

```
# Loading the third CDR dataset and getting a 
# a preview of its layout
#
url3 = "http://bit.ly/Telcom_dataset3"

cdr509 = pd.read_csv(url3)
cdr509.head()
```

**Renaming columns in the CDR Datasets**

```
# Renaming the column names in cdr507
#
cdr507.rename({"DATETIME": "DATE_TIME",  
               "PRODUTC": "PRODUCT"},  
          axis = "columns", inplace = True)

cdr507.head()
```

```
# Renaming the column names in cdr508
#
cdr508.rename({"DW_A_NUMBER": "DW_A_NUMBER_INT",  
               "DW_B_NUMBER": "DW_B_NUMBER_INT"},  
          axis = "columns", inplace = True)

cdr508.head()
```

```
# Renaming the column names in cdr509
#
cdr509.rename({"SIET_ID" : "SITE_ID",
               "CELLID" : "CELL_ID"},  
          axis = "columns", inplace = True)

cdr509.head()
```

**CLEANING THE CELLS_GEO DATASET**


```
# splitting the cells_geo dataset columns
# on the delimiter ';'
#
cells_geosplit = cells_geo[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].apply(
    lambda x : pd.Series([i for i in x.lower().split(';')]))

cells_geosplit.head()
```

```
# renaming the columns in cells_geosplit
# 
cells_geosplit.rename({ 1: "VILLES",  
                        2: "STATUS",
                        3 : "LOCALISATION",
                        4 : "DECOUP ZONE",
                        5 : "ZONE NAME",
                        6 : "LONGITUDE",
                        7 : "LATITUDE",
                        8 : "REGION",
                        9 : "AREA",
                        10 : "CELL_ID",
                        11 : "SITE_CODE"},  
                       axis = "columns", inplace = True)

cells_geosplit.head()
```

```
# Getting the unique values in the DECOUP ZONE column
#
list(cells_geosplit['DECOUP ZONE'].unique())
```

```
# Dropping the double quotations in the DECOUP ZONE
# entries
#
def decoup(decoup_str):
    return decoup_str.replace('""',"").replace('"abidjan_est"', 'abidjan_est').replace('"abidjan_nord"','abidjan_nord').replace(
        '"abidjan_centre"','abidjan_centre')

cells_geosplit['DECOUP ZONE'] = cells_geosplit['DECOUP ZONE'].map(decoup)
cells_geosplit.head()
```

```
# Getting the unique values in the ZONE NAME column
#
list(cells_geosplit['ZONE NAME'].unique())
```

```
# Dropping the double quotations in the ZONE NAME
# entries
#
def zone(zone_str):
    return zone_str.replace('""',"").replace('"assinie"', 'assinie').replace('"krikorea"','krikorea')

cells_geosplit['ZONE NAME'] = cells_geosplit['ZONE NAME'].map(zone)
cells_geosplit.head()
```

```
# Converting longitude, latitude
# column entries to floats
#
cells_geosplit.LONGITUDE = cells_geosplit.LONGITUDE.astype('float64')
cells_geosplit.LATITUDE = cells_geosplit.LATITUDE.astype('float64')

cells_geosplit.info()
```

```
# Checking for any missing values
#
cells_geosplit.isnull().sum()
```

DEALING WITH THE MISSING ENTRIES

```
# converting all spaces in the dataframe 
# to null values
#
cells_geosplit = cells_geosplit.replace(r'^\s*$', np.nan, regex=True)

cells_geosplit
```

```
# Replacing all the null values with 0
#
cells_geosplit = cells_geosplit.fillna(0)
cells_geosplit.head()
```

**MERGING OUR DATASETS AND DOING MORE CLEANING**

```
# Merging our CDR datasets
#
CDR = pd.concat((cdr507, cdr508, cdr509),
                axis = 0, ignore_index = True)
CDR
```

```
# Checking for missing values in CDR
#
CDR.isnull().sum()
```

```
# Replacing all the null values with 0 in CDR
#
CDR['SITE_ID'] = CDR['SITE_ID'].fillna(0)

CDR.isnull().sum()
```

```
# Checking for any duplicate files in the CDR dataset
# and dropping them
#
CDR.duplicated()
CDR = CDR.drop_duplicates(ignore_index= True)

CDR
```

```
# Dropping all columns that are not needed for analysis in the 
# CDR dataset
#
CDR = CDR.drop(['COUNTRY_A', 'COUNTRY_B', 'CELL_ON_SITE'], axis=1)
CDR
```
